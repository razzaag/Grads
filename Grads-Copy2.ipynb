```python
# allow inline viewing
%matplotlib inline
# import libraries
import matplotlib
from bokeh.plotting import figure, show, ColumnDataSource
from bokeh.io import output_notebook
from bokeh.models import HoverTool
import pandas as pd
import corner
```

```python
# read file
df = pd.read_csv('recent-grads.csv')
```

```python
x = df['Rank'][::-1] + 1
df = df.sort_values(['Median'])
y = df['Median']
source=ColumnDataSource(df)
```

```python
output_notebook()
```

```python
list(df)
```

```python
hover = HoverTool(tooltips=[("Rank", "@Rank"),("Major", "@Major"),])
```

```python
p = figure(title="Example", x_axis_label='Rank', y_axis_label='Median Income ($)', tools="")
p.line(x, y, line_dash="4 4", line_width=1, color='gray', source=source)
cr = p.circle(x, y, size=10,
                fill_color="grey", hover_fill_color="firebrick",
                fill_alpha=0.05, hover_alpha=0.3,
                line_color=None, hover_line_color="white", source=source)
p.add_tools(HoverTool(tooltips=[("Major", "@Major"),("Rank", "@Rank"),("Median", "@Median"),], renderers=[cr], mode='mouse'))
show(p)
```
